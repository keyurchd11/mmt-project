```python
import pandas as pd

# Read in the data
df = pd.read_csv('data.csv')


# calculate the big5 scores for each respondent

# Openness

extraversion_plus = [1,11,16,26,36]
extraversion_minus = [6,21,31]
agreeableness_plus = [7,17,22,32,42]
agreeableness_minus = [2,12,27,37]
conscientiousness_plus = [3,13,28,33,38]
conscientiousness_minus = [8,18,23,43]
neuroticism_plus = [4,14,19,29,39]
neuroticism_minus = [9,24,34]
openness_plus = [5,10,15,20,25,30,40,44]
openness_minus = [35,41]


# Create a new column for each big5 score
df['extraversion'] = 0
df['agreeableness'] = 0
df['conscientiousness'] = 0
df['neuroticism'] = 0
df['openness'] = 0

def isNumber(s):
    try:
        int(s)
        return True
    except ValueError:
        return False

# Loop through each row in the dataframe
for index, row in df.iterrows():
    # Loop through each column in the dataframe
    for column in df:
        columnNum = column.lstrip('Q')
        if not isNumber(columnNum):
            continue
        columnNum = int(columnNum)
        if columnNum in extraversion_plus:
            df.loc[index, 'extraversion'] += row[column]
        elif columnNum in extraversion_minus:
            df.loc[index, 'extraversion'] -= row[column]
        elif columnNum in agreeableness_plus:
            df.loc[index, 'agreeableness'] += row[column]
        elif columnNum in agreeableness_minus:
            df.loc[index, 'agreeableness'] -= row[column]
        elif columnNum in conscientiousness_plus:
            df.loc[index, 'conscientiousness'] += row[column]
        elif columnNum in conscientiousness_minus:
            df.loc[index, 'conscientiousness'] -= row[column]
        elif columnNum in neuroticism_plus:
            df.loc[index, 'neuroticism'] += row[column]
        elif columnNum in neuroticism_minus:
            df.loc[index, 'neuroticism'] -= row[column]
        elif columnNum in openness_plus:
            df.loc[index, 'openness'] += row[column]
        elif columnNum in openness_minus:
            df.loc[index, 'openness'] -= row[column]
        
display(df)
```

```python
# get song data from uniqueSongsWithEmotionScores.csv
song_df = pd.read_csv('uniqueSongsWithEmotionScores.csv')
# drop NAN
song_df = song_df.dropna()
display(song_df)
```

```python
# for each song of the respondent, get the sum of the emotion scores

# create a new column for each emotion score
df['anger'] = 0
df['anticipation'] = 0
df['disgust'] = 0
df['fear'] = 0
df['joy'] = 0
df['sadness'] = 0
df['surprise'] = 0
df['trust'] = 0

emotions = ['anger', 'anticipation', 'disgust', 'fear', 'joy', 'sadness', 'surprise', 'trust']
pos_tags = ['Coordinate conjunction', 'Preposition or subordinating conjunction', 'Adjective', 'Noun', 'Pronoun', 'Adverb', 'TO', 'Interjection', 'Verb']

for pos in pos_tags:
    df[pos] = 0

for index, row in df.iterrows():
    # Loop through each column in the dataframe
    for column in df:   
        if column.startswith('song'):
            songID = row[column]
            if songID == 0:
                continue
            song = song_df.loc[song_df['id'] == songID]
            for emotion in emotions:
                df.loc[index, emotion] += song[emotion].values[0]
            for pos in pos_tags:
                df.loc[index, pos] += song[pos].values[0]
```

```python
df2 = df
df2.drop(columns=df2.columns[:64],axis=1, inplace=True)
```
